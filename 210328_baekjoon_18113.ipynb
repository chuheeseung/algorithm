# 백준 18113 그르다 김가놈
### https://www.acmicpc.net/problem/18113

```python
n, k, m = map(int, input().split())
l = []

for _ in range(n):
    l.append(int(input()))

result = -1

def chopping(mid):
    cnt = 0
    
    for gb in l :
        if gb <= k :
            continue
        elif gb < 2 * k:
            cnt += (gb - k) // mid
        elif gb > 2 * k:
            cnt += (gb - (2 * k)) // mid
    return cnt;

def binSearch(low, high):
    while(low <= high):
        mid = (low + high) // 2
        
        if mid == 0:
            high = mid - 1
            break
        
        if chopping(mid) < m:
            high = mid - 1
        else:
            low = mid + 1
    return high

if max(l) > k:
    result = binSearch(0, max(l) - k)

print(result)    
```
