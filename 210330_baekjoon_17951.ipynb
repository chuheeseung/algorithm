# 백준 17951
### https://www.acmicpc.net/problem/17951

```python
# import sys
# Paper = sys.stdin.readline #속도 조절용

#n : 시험지의 개수
#k : 시험지를 나눌 그룹의 수
n, k = list(map(int, input().split()))
score = [] #각 시험지마다 맞은 문제의 개수
score = list(map(int, (input().split())))
# print(score)

def calculate(num): #점수 계산 함수
    group = 1 #그룹 개수
    score_sum = 0 #점수를 더한 변수
    
    for i in range(n):
        score_sum += score[i] #x에 있는 변수만큼 더함
        
        #그룹 내 점수가 n 이상이 되면 그룹 자르기
        if(score_sum > num):
            group += 1 #그룹 개수 증가
            score_sum = 0 #점수 초기화
            
    return group <= k #그룹 개수가 k개 보다 작거나 같을 때 반환        

minimum = 0
maximum = max(score) * n + 1

while minimum + 1 < maximum:
    mid = (minimum + maximum) // 2 #최대 점수
    
    if (calculate(mid)): #점수 계산함수를 통해 계산
        #그룹의 개수가 k보다 작은 경우 
        maximum = mid #최대 점수를 mid로 내리고
    else:
        minimum = mid #최소 점수를 mid로 올린다
        
print(maximum) #현수가 받을 수 있는 최대 점수 출력        
```
