# 백준 20302

###### 디저트 고르는 상원이
###### 정수의 곱셈,나눗셈 수식
###### 결과가 정수 - 민트초코 / 유리수 - 치약
###### 어떤 디저트를 먹을지 맞춰보자

###### n: 수식을 이루는 수 개수
###### 수식: 정수, 연산자(*, /) 공백으로 구분


- 모든 합성수는 소인수분해를 통해 소수의 곱으로 표현할 수 있고, 그 개수들을 카운트 하기 위해 소수들의 개수를 담는 배열을 사용한다. 또한 정수, 유리수 판별에 음수 여부는 관계 없으므로 편의를 위해 모두 양수로 바꿔준다
- 곱하는 수는 소인수분해해서 표현된 소수들의 개수를 올려주고, 나누는 수는 소인수분해하여 표현된 소수들의 개수를 빼주어 그 개수들을 관리할 수 있다. (곱하는 수 중 0이 존재한다면 항상 결과는 0이므로 이는 따로 처리해준다)
- 모든 수를 소인수분해 하고 나서 [2, 1e5] 소수 배열을 순회하는데 그 값이 음수인 경우 분모에 수가 남게 되는 것이므로 이는 유리수라고 정할 수 있고 그 외는 정수라고 출력하면 된다

```python
import math

n = int(input())
num = [] #숫자만 따로 저장할 배열
oper = [] #연산자만 따로 저장할 배열
ent = [] #전체를 입력받을 배열
ent = list(input().split())

zero = False #0 존재여부 확인용
ratNum = False #유리수 여부 확인용
max = int(1e5) #최댓값 100 000
primeNum = [0] * (max + 1) #소수 개수 저장할 배열

#전체 입력받은 배열에서 숫자와 연산자를 분리해서 저장
for i in range(len(ent)):
    if i % 2 == 0:
        num.append(int(ent[i]))
    else:
        oper.append(ent[i])

#소인수분해 함수
#곱셈 연산자 다음에 온 수인 경우: 해당 소수의 갯수 하나씩 더함
#소인수분해를 통해서 소수의 개수 증감(합성수)
#해당 연산자 다음에 온 수의 소수 개수를 증감해야 함
#곱셈 - 소수의 개수를 추가
def mul(n):
    root = int(math.sqrt(n)) #math.sqrt :d 제곱근 구하는 함수
    #다음에 온 소수의 개수를 증가
    for idx in range(2, root + 1):
        #해당되는 숫자의 소수 개수를 증가시키기 위해
        while((n % idx) == 0):
            n //= idx #합성수를 위한 나눗셈
            primeNum[idx] += 1 #소수의 개수 증가
    if (n > 1): #소인수가 1개인 경우
        primeNum[n] += 1 #소수 개수 증가
        

#나눗셈 연산자 다음에 온 수인 경우: 해당 소수의 갯수 하나씩 뺌
#나눗셈 - 소수의 개수 감소
def div(n): 
    root = int(math.sqrt(n))
    #다음에 온 수의 소수 개수를 감소
    for idx in range(2, root + 1):
        #해당되는 숫자의 소수 개수를 감소시키기 위해
        while((n % idx) == 0):
            n //= idx #합성수를 위한 나눗셈
            primeNum[idx] -= 1 #소수 개수 감소
    if (n > 1): #소인수가 1개인 경우
        primeNum[n] -= 1 #소수 개수 감소
        
for i in range(len(num)):
    #첫번째는 앞에 연산자가 없으므로 따로 계산
    if i == 0:
        #정수와 유리수 판별에 음수는 관계 없으니까 양수로 변환
        #모든 정수는 -100000 이상 100000 이하이므로 다음 조건이 필요
        if num[i] < 0:
            num[i] *= -1 #음수는 정수/유리수 판별에 소용이 없어서 양수로
        #0이 있으면 무조건 계산결과는 0(=정수)이 됨
        elif num[i] == 0:
            zero = True
            break
        else: #첫번째 숫자이므로 소수의 개수 증가
            mul(num[i])
    else:
        #정수와 유리수 판별에 음수는 관계 없으므로 편의상 양수로 변환
        if num[i] < 0:
            num[i] *= -1
        #0인 경우는 정수    
        if num[i] == 0:
            zero = True
            break
            
        #나눗셈 다음에 오는 수면 div 호출, 곱셈이면 mul 호출
        if oper[i - 1] == '/':
            div(num[i])
        elif oper[i - 1] == '*':
            mul(num[i])
            
if zero: #정수면
    print("mint chocolate") 
else:
    #모든 수를 소인수분해하고 2 ~ 100000 에서 소수 배열을 순회
    #이때 그 값이 음수인 경우는 분모에 수가 남으므로 유리수
    #그 외에는 정수
    for i in range(2, max + 1):
        if primeNum[i] < 0: #음수인 경우에는 유리수
            ratNum = True
    
    if ratNum: #유리수
        print("toothpaste")
    else: #정수
        print("mint chocolate")
    
    
```

### 에라토스테네스의 체 
### : 범위에서 합성수를 지우는 방식으로 소수를 찾는 방법

##### 1. 1은 제거
##### 2. 지워지지 않은 수 중 제일 작은 2를 소수로 채택하고, 나머지 2의 배수를 모두 지운다
##### 3. 지워지지 않은 수 중 제일 작은 3을 소수로 채택하고, 나머지 3의 배수를 모두 지운다
##### 4. 지워지지 않은 수 중 제일 작은 5를 소수로 채택하고, 나머지 5의 배수를 모두 지운다
##### 5. (반복)

```python
n = 100

#소수인지 검사하는 함수 isPrime()
def isPrime(a):
    if(a < 2):
        return False
    
    for i in range(2, a):
        if(a % i == 0):
            return False
    return True

for i in range(n + 1):
    if(isPrime(i)):
        print(i)
```

```python
#에라토스테네스의 체
n = 100
a = [False, False] + [True] * (n - 1)
primes = []

for i in range(2, n + 1):
    if a[i]:
        primes.append(i)
        for j in range(2 * i, n + 1, i):
            a[j] = False
print(primes)            
```
