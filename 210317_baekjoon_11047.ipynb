# 백준 11047 동전 0 
### https://www.acmicpc.net/problem/11047

```python
n, k = map(int, input().split()) #n, k 입력 받음
coin_list = [] #동전의 가치 저장할 배열 선언
coin_count = 0 #동전의 개수를 셀 변수

coin_list = [int(input()) for _ in range(n)]
coin_list = coin_list[::-1] #내림차순으로 변경

# print(coin_list)

for i in coin_list :
    count = k // i
    coin_count += count
    k -= count * i
    
    if k == 0 : 
        break

print(coin_count)        
```
