# 백준 7576 토마토
### https://www.acmicpc.net/problem/7576

하루가 지나면 익토에 인접한 토마토가 익음
인접 : 상하좌우
혼자 저절로 익는 경우 없음

구) 창고에 보관된 토마토들이 며칠이 지나면 익게 되는지 최소 일수

입) 격자상자 크기, 토마토 정보
m: 가로 수 / n: 세로 수
1: 익토 / 0: 안익토 / -1: 없음

출) 최소 일수
0: 모든 토마토가 익어있는 상태
-1: 모든 토마토가 익지는 못하는 상황


```python
import sys
from collections import deque

m, n = map(int, input().split())

tomato = []
for _ in range(n):
    tomato.append(list(map(int, input().split())))

count = 0
dx = [0, 0, -1, 1]
dy = [1, -1, 0, 0]
coord = deque() #큐 선언

for i in range(n):
    for j in range(m):
        if tomato[i][j] == 1:
            coord.append([i, j])

while len(coord) != 0:
    count += 1
    for _ in range(len(coord)):
        x, y = coord.popleft()
        
        for i in range(4):
            new_x = x + dx[i]
            new_y = y + dy[i]
            
            if(0 <= new_x < n) and (0 <= new_y < m) and tomato[new_x][new_y] == 0:
                tomato[new_x][new_y] = 1
                coord.append([new_x, new_y])
                
count -= 1

for i in tomato:
    if 0 in i:
        count -= -1
        break
        
print(count)        
```
