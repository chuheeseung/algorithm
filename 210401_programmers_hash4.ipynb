````python
def solution(genres, plays):
    answer = []
    genre_dic = {}
    total_list = {}
    
    for i in range(len(genres)):
        if genres[i] in genre_dic:
            genre_dic[genres[i]] += plays[i]
            total_list[genres[i]].append((plays[i], i))
        else:
            genre_dic[genres[i]] = plays[i]
            total_list[genres[i]] = [(plays[i], i)]
            
    genre_dic = sorted(genre_dic.items(), key = lambda x : x[1], reverse = True)
    #sorted(정렬데이터, key파라미터, reverse파라미터)
    #key 값을 기준으로 비교를 해서 정렬하겠다는 뜻
    #reverse = False : 오름차순, True : 내림차순
    
    for i in total_list:
        total_list[i] = sorted(total_list[i], key = lambda x : x[0], reverse = True)[:2] #길이 2개로 정해져서 자른건가봄
#         print(total_list)
     
    while len(genre_dic) > 0:
        genre_max = genre_dic.pop(0)
        
        for t in total_list: #t가 지금 classic이냐 pop이냐로 돌아가고 있는거임
            if t == genre_max[0]:
                if len(total_list[t]) > 1: #여기서 리스트로 찝는 부분이 이해가 안된다
                    answer.append(total_list[t][0][1])
                    answer.append(total_list[t][1][1])
                else:
                    answer.append(total_list[t][0][1])
    
    return answer
````

````python
genres = ["classic", "pop", "classic", "classic", "pop"]
plays = [500, 600, 150, 800, 2500]
````

````python
solution(genres, plays)
````

### 다른 풀이

````python
def solution1(genres, plays):
    answer = []
    d = {e : [] for e in set(genres)} #{'classic': [], 'pop': []}
    
    for e in zip(genres, plays, range(len(plays))): 
        #e : ('classic', 500, 0)('pop', 600, 1)('classic', 150, 2)('classic', 800, 3)('pop', 2500, 4)
        d[e[0]].append([e[1], e[2]]) #d : {'classic': [[500, 0], [150, 2], [800, 3]], 'pop': [[600, 1], [2500, 4]]}
        
    genreSort = sorted(list(d.keys()), key = lambda x : sum(map(lambda y : y[0], d[x])), reverse = True)
    #genreSort : ['pop', 'classic']
    '''
    Q. lambda 부분 이해가 안됨
    A. x: list(d.keys())의 원소 / map>lambda>y: d[x]의 원소
       x(genre)를 받아서 d[x]에서 play된 수만 추출, 그 값들을 모두 합한 값을 key로 sort
    '''
    
    for g in genreSort:
        temp = [e[1] for e in sorted(d[g], key = lambda x : (x[0], -x[1]), reverse = True)] 
        #temp : [4, 1], [3, 0, 2]
        '''
        Q. lambda 부분에서 -x[1] 이해가 안됨
        '''
        answer += temp[:min(len(temp), 2)]
    
    return answer
````

````python
solution1(genres, plays)
````

1. dictionary에 값을 넣을 때 이렇게 할 필요가 없음
``` python
dict = {}
if key in dict:
    dict[key] += temp
```

``` python
from collections import defaultdict
dict = defaultdict(int) #디폴트 값이 int인 딕셔너리 (list, set 등 가능)

```

2. sorted(정렬할 데이터, key 파라미터, reverse 파라미터)
##### key 파라미터 : 정렬할 때 key를 기준으로 비교해서 정렬하겠다는 뜻
##### reverse 파라미터 : True 내림차순, False 오름차순
##### 리스트.sort()와 sorted(리스트)의 차이점 : .sort()는 본체를 정렬, sorted()는 정렬한 새로운 리스트를 반환

3. 왜 출력결과에서 2가 없는거지?
##### => 제한 사항에 genres와 plays의 길이는 같다고 나와 있어서 길이가 더 짧은 plays의 길이에 맞춰진 것
#####    제일 짧은 p : (150, 2)가 없어진거임
